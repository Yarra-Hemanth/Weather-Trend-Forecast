## Performing Augmented Dickey-Fuller (ADF) test

```python
from statsmodels.tsa.stattools import adfuller


adf_test = adfuller(df['temperature_celsius'])
print(f"ADF Statistic: {adf_test[0]}")
print(f"P-value: {adf_test[1]}")

df['temp_diff'] = df['temperature_celsius'].diff()
df['temp_diff'].fillna(df['temp_diff'].mean(), inplace=True)
```

```python
df
```

## Test, Train Split

```python
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]
train.head()
```

Looking for significant spikes in the plots:

PACF: First significant spike at lag 1 → p = 1

ACF: First significant spike at lag 1 → q = 1

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(211)
plot_acf(df['temperature_celsius'].diff().dropna(), lags=30, ax=plt.gca())
plt.subplot(212)
plot_pacf(df['temperature_celsius'].diff().dropna(), lags=30, ax=plt.gca())
plt.show()
```

## Temperature Forecasting with SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import LabelEncoder

model = SARIMAX(train['temperature_celsius'],
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1,30),
                exog=train.drop(columns=['temperature_celsius']),
                enforce_stationarity=False, enforce_invertibility=False)

model_fit = model.fit()

forecast = model_fit.predict(start=len(train), end=len(df)-1,
                             exog=test.drop(columns=['temperature_celsius']), dynamic=True)

plt.figure(figsize=(14, 5))
plt.plot(train.index, train['temperature_celsius'], label="Train Data", color="blue")
plt.plot(test.index, test['temperature_celsius'], label="Actual Data", color="orange")
plt.plot(test.index, forecast, label="SARIMA Forecast", color="red", linestyle="dashed")
plt.legend()
plt.title("SARIMA Forecast for Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.show()
```

## Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(test['temperature_celsius'], forecast)
mse = mean_squared_error(test['temperature_celsius'], forecast)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (RMSE): {mse:.2f}")
```

## Temperature Forecasting using Prophet

```python
from prophet import Prophet

df = df_country

df = df.reset_index()
df = df.rename(columns={'last_updated': 'ds', 'temperature_celsius': 'y'})
df['ds'] = pd.to_datetime(df['ds'])

prophet_df = df[['ds', 'y']]

model = Prophet()
model.fit(prophet_df)

future = model.make_future_dataframe(periods=30, freq='D')

forecast = model.predict(future)

forecast.to_csv("weather_forecast.csv", index=False)
print("Forecasting complete! Results saved to weather_forecast.csv")
```

## Plotting Forecast

```python
fig1 = model.plot(forecast)
fig2 = model.plot_components(forecast)
plt.show()
```

## Visualization(Actual vs Forecasted Temperature)

```python
plt.figure(figsize=(12, 6))
plt.plot(df['ds'], df['y'], label='Actual', color='blue')
plt.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='red', linestyle='dashed')
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='pink', alpha=0.3)

plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Actual vs Forecasted Temperature')
plt.legend()
plt.grid()
plt.show()
```

## Evaluating Prophet Forecast

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

evaluation_df = df.merge(forecast[['ds', 'yhat']], on='ds', how='left').dropna()

mae = mean_absolute_error(evaluation_df['y'], evaluation_df['yhat'])
mse = np.sqrt(mean_squared_error(evaluation_df['y'], evaluation_df['yhat']))

print(f"Mean Absolute Error (MAE): {mae:.3f}")
print(f"Mean Squared Error (RMSE): {mse:.3f}")
```
